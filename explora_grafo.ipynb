```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def dibuja_distr_grado(G):
    degree_counts = nx.degree_histogram(G)

    # Grafica la distribución de grado
    #plt.figure(figsize=(10, 6))
    fig, ax = plt.subplots()
    ax.set_yscale("log")
    ax.bar(range(len(degree_counts)), degree_counts)
    plt.xlabel("Grado")
    plt.ylabel("Número de Nodos")
    plt.title("Distribución de Grado")
    plt.grid()
    plt.show()
```

```python
file_path = "Red Diccionario\Oxford_Dict_Graph.gml"
G = nx.read_gml(file_path)
```

```python
file_path_adj = "Red Diccionario\Oxford_Dict_Graph_Adj.gml"
G_adj = nx.read_gml(file_path_adj)
```

```python
dibuja_distr_grado(G)
```

```python
dibuja_distr_grado(G_adj)
```
